## 8. Exploring a Numeric Variable

```python
col = numeric_cols[0]  # change this if you want

data[col].hist(bins=20)
plt.xlabel(col)
plt.ylabel('Frequency')
plt.title(f'Distribution of {col}')
plt.show()
```

## 9. Exploring a Categorical Variable
