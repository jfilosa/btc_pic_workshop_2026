# Lesson 2: Logic

Coding logic is how you can use Python to make decisions. Some common logic you will come across are "if" "else" statements. These are your best friends!

Combined with logic operators (or, and, less than, greater than) it is a powerful programming tool.

## Exercise 2.1: Logic

```python
age= 12

if age<10:
    print("Your age is single digits!")
elif age>=10:
    print("Your age is not single digits")
```
