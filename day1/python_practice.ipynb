## ðŸŽ¯ Final Exercise: Class Grade Calculator

### Why this is useful

This project shows how Python can help calculate **final grades**, something students actually care about and understand.

### Scenario

A class has **four assignments** with different weights:

* Homework: **20%**
* Quiz: **20%**
* Project: **30%**
* Final Exam: **30%**

Each score is out of 100.

### Goal

Write a program that:

1. Asks the user for their score on each assignment
2. Calculates the weighted average
3. Prints the final numeric grade
4. Prints the **letter grade** (A, B, C, D, or F)

### Starter Code

```python
homework = float(input("Enter your homework grade (0-100): "))
quiz = float(input("Enter your quiz grade (0-100): "))
project = float(input("Enter your project grade (0-100): "))
final_exam = float(input("Enter your final exam grade (0-100): "))

final_grade = (
    homework * 0.20 +
    quiz * 0.20 +
    project * 0.30 +
    final_exam * 0.30
)

print("Final grade:", final_grade)

if final_grade >= 90:
    print("Letter grade: A")
elif final_grade >= 80:
    print("Letter grade: B")
elif final_grade >= 70:
    print("Letter grade: C")
elif final_grade >= 60:
    print("Letter grade: D")
else:
    print("Letter grade: F")
```
